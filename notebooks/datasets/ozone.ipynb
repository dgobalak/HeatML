```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("Aerosol Summary.csv")
df
```

```python
df1 = df.loc[(df['Latitude (degree)'] == 37.3223) & (df['Time (TAI 1993)']==278427000.0)]
df1
```

```python
df2 = df.loc[(df['Latitude (degree)'] == 37.3223) & (df['Time (TAI 1993)']==352677000.0)]
df2
```

```python
df1 = df1.drop(df1[df1["Aerosol extinction coefficient (km^-1)"] < -10].index)
df2 = df2.drop(df2[df2["Aerosol extinction coefficient (km^-1)"] < -10].index)
```

```python
import plotly.express as px
fig1 = px.line(df1, x=" Altitude (km)", y="Aerosol extinction coefficient (km^-1)", title="Earlier Measurements")
fig2 = px.line(df2, x=" Altitude (km)", y="Aerosol extinction coefficient (km^-1)", title="Later Measurements")
```
